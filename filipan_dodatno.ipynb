### 1. zadatak

Odredite optimalnu složenost modela linearne regresije s polinomijalnim preslikavanjem pomoću metode unakrsne validacije. Stvorite $30$ točaka za učenje i $15$ točaka za provjeru pomoću funkcije `generate_data`. Obavezno komentirajte kako ste odredili optimalan stupanj preslikavnja i prikažite graf pogreške u ovisnosti o složonesti modela na oba skupa za učenje.

```python
def generate_data(n):
    x = np.random.uniform(-3, 3, size=(n))
    y = x**3 * np.sin(x) + 2 * x + np.random.normal(0, 1, size=n) + 3
    return x, y


x, y = generate_data(100)
plt.scatter(x,y)

```
