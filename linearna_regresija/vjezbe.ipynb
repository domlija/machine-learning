```python
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
N = 100

X = np.random.uniform(0, 5, size=N)

# y = 2*x + 1
y =  2 * X + 1 + np.random.normal(0, 1, size=N)
# for i in range(len(X)): 
#   y[i] = 2* X[i] + 1 + noise[i]

plt.scatter(X,y)

ls = np.linspace(0,5,100)
plt.plot(ls, ls * 2 + 1)
plt.show()
```

```python
model = LinearRegression()  #https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

model.fit(X.reshape(-1,1), y) # [1,2,3] -> [[1], [2], [3]]
y_pred = model.predict(X.reshape(-1,1))

plt.scatter(X, y)
ls = np.linspace(0, 5, 100)
ls = ls.reshape(-1,1)

ls_pred = model.predict(ls)
plt.plot(ls, ls_pred, label='prediction')

#ls = np.linspace(0,5,100)
plt.plot(ls, ls * 2 + 1, label='real')
plt.legend()
plt.show()


```

```python
N = 10

X = np.random.uniform(0, 5, size=N)
X = np.linspace(-5,5,N)

y =  X**2 + 1 #+ np.random.normal(0, 1, size=N)


plt.scatter(X,y)

model = LinearRegression()

model.fit(X.reshape(-1,1),y)
y_pred = model.predict(X.reshape(-1,1))

print(f'MSE is {mean_squared_error(y, y_pred)}')
print(f'r2_score {r2_score(y, y_pred)}')

ls = np.linspace(-5, 5, 100)
ls = ls.reshape(-1,1)

ls_pred = model.predict(ls)
plt.plot(ls, ls_pred, label='prediction')
plt.show()

```

```python
N = 10

X = np.random.uniform(0, 5, size=N)
X = np.linspace(0,5,N)

y =  2 * X + 1 + np.random.normal(0, 1, size=N)



plt.scatter(X,y)

model = LinearRegression()

model.fit(X.reshape(-1,1),y)

y_pred = model.predict(X.reshape(-1,1))

print(f'MSE is {mean_squared_error(y, y_pred)}')
print(f'r2_score {r2_score(y, y_pred)}')
print(f'w_1: {model.coef_} w_0: {model.intercept_}')

ls = np.linspace(0, 5, 100)
ls = ls.reshape(-1,1)

ls_pred = model.predict(ls)
plt.plot(ls, ls_pred, label='prediction')

# h = w1 * x
# a inae je oblika h = w0 + w1 * x

```
